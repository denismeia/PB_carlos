```python
a_f,m = xu.TifToArray('alb_f.tif')
a_f = a_f.flatten()
a_f = np.where(a_f==-99999.0,np.nan,a_f)

a_n,m = xu.TifToArray('alb_n.tif')
a_n = a_n.flatten()
a_n = np.where(a_n==-99999.0,np.nan,a_n)

l_f,m = xu.TifToArray('lai_f.tif')
l_f = l_f.flatten()
l_f = np.where(l_f==-99999.0,np.nan,l_f)

l_n,m = xu.TifToArray('lai_n.tif')
l_n = l_n.flatten()
l_n = np.where(l_n==-99999.0,np.nan,l_n)

e_f,m = xu.TifToArray('et_f.tif')
e_f = e_f.flatten()
e_f = np.where(e_f==-99999.0,np.nan,e_f)

e_n,m = xu.TifToArray('et_n.tif')
e_n = e_n.flatten()
e_n = np.where(e_n==-99999.0,np.nan,e_n)
```

```python
from scipy.stats import kendalltau
```

```python
i = 100
sn.set(style="ticks")

sn.jointplot(l_f[::i],e_f[::i],kind="kde", size=4, space=0, color='grey')
plt.title('LAI x ET',y=0, x=-3.2)
plt.savefig('scatter_LAI-ET.pdf',bbox_inches='tight')

sn.jointplot(l_f[::i],a_f[::i],kind="kde", size=4, space=0, color='grey')
plt.title('LAI x Albedo',y=0, x=-3.2)
plt.savefig('scatter_LAI-albedo.pdf',bbox_inches='tight')

sn.jointplot(a_f[::i],e_f[::i],kind="kde", size=4, space=0, color='grey')
plt.title('Albedo x ET',y=0, x=-3.2)
plt.savefig('scatter_Albedo-ET.pdf',bbox_inches='tight')

plt.show()
```
