```python
from glob import glob
import numpy as np
import xuleta as xu
import os

fmyd = "/media/denis/seagate/PB/MODIS/myd16a2/"
fmod = "/media/denis/seagate/PB/MODIS/mod16a2/"
fout = "/media/denis/seagate/PB/MODIS/mxd16a2"

if not os.path.exists(fout):
    os.makedirs(fout)

lmyd = xu.listfiles(fmyd, extension="*.tif", full=False)
lmod = xu.listfiles(fmod, extension="*.tif", full=False)
linter = sorted(list(set(lmyd).intersection(lmod)))
```

```python
#predomina myd
for i in linter:
    myd,m = xu.TifToArray(fmyd+i)
    mod,m = xu.TifToArray(fmod+i)
    myd = np.where(myd > 30000, np.nan, myd)
    mod = np.where(mod > 30000, np.nan, mod)
    mxd = np.where(mod>myd, mod, myd)
    mxd = np.where(mxd==np.nan,32767,mxd)
    xu.ArrayToTif(mxd,i,fout,m,Type=1)
```
